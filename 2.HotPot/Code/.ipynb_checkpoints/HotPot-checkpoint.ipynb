### 3. 统计火锅团购商家店名高频词，并绘制出词云

```python
# 导入词云制作库wordcloud和中文分词库jieba
import re
import jieba
import wordcloud
import collections
# 导入imageio库中的imread函数，并用这个函数读取本地图片，作为词云形状图片
import imageio

mk = imageio.imread("hotpot.jpg")
# 设置字体路径
font_path=r"D:\MyProjects\Fudan\Data Mining\2.HotPot\msyh.ttf"
# 构建并配置词云对象w，注意要加stopwords集合参数，将不想展示在词云中的词放在stopwords集合里，这里去掉“曹操”和“孔明”两个词
w = wordcloud.WordCloud(width=1000,
                        height=700,
                        background_color='white',
                        font_path=font_path,
                        mask=mk,
                        scale=15,
                        stopwords={'区','路','重庆','北京','火锅','小','小火锅',
                                   '焖','煲','涮','串串','小区','分店','路店','店','老','锅',
                                   '门店','火锅店','成都'})

# 提取文本
myword = list(data["商家店名"])
myword = " ".join(myword)

# 文本预处理
# 定义正则表达式匹配模式
pattern = re.compile(u'\t|\n|\.|-|:|;|\)|\(|\?|"')
# 将符合模式的字符去除
string_data = re.sub(pattern, '',myword)

# 对文本进行中文分词，得到string
txtlist = jieba.lcut(myword)
string = " ".join(txtlist)

# 原始数据处理（定义需要去掉的词汇与字符）
object_list = []
remove_words = ['区','路','重庆','北京','火锅','小','小火锅','焖','煲','涮','串串','小区','分店','路店',
                '店','老','锅','(',')',' '] # 自定义去除词库

for word in txtlist: # 循环读出每个分词
    if word not in remove_words: # 如果不在去除词库中
        object_list.append(word) # 分词追加到列表

# 词频统计
# 对分词做词频统计
word_counts = collections.Counter(object_list)
# 获取前10最高频的词
word_counts_top10 = word_counts.most_common(10)
word_counts_top10_list = []
for i in range(10):
    word_counts_top10_list.append(word_counts_top10[i][0])

# 将string变量传入w的generate()方法，给词云输入文字
w.generate(string)
# 画出词云图
plt.figure(figsize=(15, 10))
#plt.subplot(121)
plt.imshow(w)
plt.axis('off')
plt.show()

# 将词云图片导出到当前文件夹
# w.to_file('wc_hotpot.jpg')
```

在以上词云图中可以清晰看到商家店名中最常见的词语包括："自助"、"私房"、"鸡公煲"、"串串香"、"清真"、"芭蕉"、"泰式"
等等，说明消费者比较偏好于"自助"、"私房"这两大主题的火锅店，反映了中国食客偏好大食量、有特色的火锅店。
从口味来讲，"鸡公煲"、"串串香"、"芭蕉/泰式"是受消费者偏爱的，意味着除了传统中华美食之外，泰国菜近年来也深受
中国食客的热爱。

### 5. 建立线性回归模型（y=季均销量)

#### 5.1 数据描述与数据预处理

1. 数据共有24个特征，其中包含7个定量变量，11个定性变量（两个水平）、6个定性变量（多个水平）
2. 定量变量处理：归一化（Normalization），由于数据在实际中都是大于零的数，所以用归一化处理要好于标准化，处理后的数据均在0到1之间。
3. 定性变量（两个水平）处理：标签化，由于这些变量只有2个水平，故将其化作0-1型变量。
4. 定性变量（多个水平）处理：化为哑变量，每一组均选定一个特征作为基准组，处理后变为20个定性变量。
5. 总结：经过处理后，变为了38个特征，其中有7个定量变量、11个定性变量（两个水平）、20个定性变量（多个水平）


```python
# 数据处理
# 定量变量
quant_var1 = ['套餐售价','团购折扣力度','营业时长','评分','评论数','在线天数']
# 定量变量：店名信息
quant_var2 = ["使用TOP20高频词的个数"]
# 定量变量汇总
quant_vars = ['套餐售价','团购折扣力度','营业时长','评分','评论数','在线天数',"使用TOP20高频词的个数"]
print('定量变量共有:',len(quant_vars),'个')

# 定性变量(两个水平)
quali_var1 = ['是否钱包新客优惠','是否立减优惠','是否节假日通用','是否需预约','每次消费团购张数限制','是否提供WIFI']
# 定性变量(两个水平)汇总
quali_vars1= ['是否钱包新客优惠','是否立减优惠','是否节假日通用','是否需预约','每次消费团购张数限制','是否提供WIFI',
              "是否包含地名","是否包含食材","是否包含口味","是否包含烹饪方式","是否包含人物"]
print('定性变量(两个水平)共有:',len(quali_vars1),'个')

# 定性变量(多个水平)
quali_var2 = ['所属行政区','每张团购券使用人数限制','停车位','包间']
# 定性变量(多个水平)汇总
quali_vars2= ['所属行政区','每张团购券使用人数限制','停车位','包间',"菜系"]
print('定性变量(多个水平)共有:',len(quali_vars2),'个')

# 定性变量：店名信息
quali_var3 = ["是否包含地名","是否包含食材","是否包含口味","是否包含烹饪方式","是否包含人物","菜系"]
print('定量变量(店名信息)共有:',1,'个(使用TOP20高频词的个数)')
print('定性变量(店名信息)(两个水平)共有:',len(quali_var3),'个')
print('定性变量(店名信息)(多个水平)共有:',1,'个(菜系)')
```

```python
data_ = data.copy()

# 对定量变量做归一化处理
data_[quant_vars] = MinMaxScaler().fit_transform(data_[quant_vars])

# 定性变量(两个水平）做0-1标签化处理
# 定义一个空字典，放入分类后的情况
quali_dict = {}
def str2num(col_name):
    le = LabelEncoder()
    le = le.fit(col_name)
    value_dict = {}
    for num,value in enumerate(list(le.classes_), start = 0):    # enumerate:返回索引、值
        value_dict[value] = num
    return value_dict

# 对每个2值分类变量进行处理
for i in quali_vars1:
    quali_dict[i] = str2num(data_[i])
    le = LabelEncoder()
    le = le.fit(data_[i])
    data_[i] = le.fit_transform(data_[i])

# 查看分类后的情况
print('定性变量(两个水平)分类后情况：')
quali_dict
```

```python
# 定性变量(多个水平）做哑变量处理
dummy1 = pd.get_dummies(data_[quali_vars2])
dummy2 = pd.get_dummies(data_[quali_vars2],drop_first=True)

mynewcol = []
for i in list(dummy1.columns):
    if i in list(dummy2.columns):
        pass
    else:
        mynewcol.append(i)

print('定性变量(多个水平)基准组分别为：\n',mynewcol)

# 将处理好的定性变量与原表格拼接，并删去原表格无用列
newdata = pd.concat([data_,dummy2],axis=1)
newdata.drop(quali_vars2,axis=1,inplace=True)
newdata.drop(["商家店名","广告栏关键词"],axis=1,inplace=True)

# 调整表格列顺序
order = [
         # 因变量【1个】
         '季均销量',
         # 定量变量【7个】
         '套餐售价','团购折扣力度','营业时长','评分','评论数','在线天数','使用TOP20高频词的个数',
         # 定性变量（2个水平）【11个】
         '是否钱包新客优惠','是否立减优惠','是否节假日通用','是否需预约','是否提供WIFI','每次消费团购张数限制',
         '是否包含地名','是否包含食材','是否包含口味','是否包含烹饪方式','是否包含人物',
         # 定性变量（多个水平）【共20个】
         # 基准组：'菜系_其他'【3个】
         '菜系_北系','菜系_川系','菜系_粤系',
         # 基准组：'所属行政区_临潼区'【9个】
         '所属行政区_新城区','所属行政区_未央区','所属行政区_灞桥区','所属行政区_碑林区','所属行政区_莲湖区','所属行政区_长安区',
         '所属行政区_阎良区','所属行政区_雁塔区','所属行政区_高新区',
         # 基准组：'每张团购券使用人数限制_不限人数'【3个】
         '每张团购券使用人数限制_限1人','每张团购券使用人数限制_限2-4人','每张团购券使用人数限制_限5-12人',
         # 基准组：'停车位_免费停车位'【3个】
         '停车位_收费停车位','停车位_无停车位','停车位_未提及',
         # 基准组：'包间_仅限大厅使用'【2个】
         '包间_均可使用','包间_未提及']
newdata = newdata[order]
```

#### 5.2 线性回归

```python
# 进行线性回归
# 使用前2000行数据作为训练集
import statsmodels.api as sm
X = newdata[newdata.columns[1:]][:2000]
Y = newdata['季均销量'][:2000]
X = sm.add_constant(X)
est = sm.OLS(Y,X).fit()
est.summary()
```

```python
# 利用训练模型进行预测
# 使用后688行数据作为训练集
X_test = newdata[newdata.columns[1:]][2000:]
Y_test = newdata['季均销量'][2000:]
X_test = sm.add_constant(X_test)
Y_hat = est.predict(X_test)

# 计算RMSE
rmse_loss = np.sqrt(np.square(Y_test - Y_hat).mean())
print('rmse_loss:', rmse_loss)
```

### 5.3 模型解读

一、 定量变量
1. 套餐售价、团购折扣力度、在线天数与季均销量成反比：首先价格与销量成反比是正常的经济现象，团购折扣力度越大（即数值越小）则销量越高，而在线天数越长效率越低，意味着越难卖的产品在线天数也就越长。
2. 营业时长. 评分. 评论数. 使用TOP20高频词的个数与季均销量成正比：营业时间越长、评分越高、评论数越多、店名中包含的高频词汇越多，则销量越大，也是容易理解的现象。

二、 定量变量（两个水平）
