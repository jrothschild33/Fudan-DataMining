```python
# 导入所需库
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.api as sm
import matplotlib.pyplot as plt
%matplotlib inline

# 设置表格显示，默认显示所有列
pd.set_option('display.max_columns', 10)
```

# 第1部分 导入数据并进行处理

```python
# 导入数据
data_file = 'Speed Dating Data.csv'
df_full = pd.read_csv(data_file, encoding='ISO-8859-1')
print('====================数据已成功导入====================')
print('文件名称：{}'.format(data_file))
print('数据大小:',df_full.shape[0],'条')
print('表格列头:',df_full.shape[1],'列')

# 将数据类别进行规整分类
personal = ['gender', 'age', 'field', 'field_cd', 'undergra',
           'mn_sat', 'tuition', 'race', 'imprace', 'imprelig', 'from',
           'zipcode', 'income', 'goal', 'date', 'go_out', 'career',
           'career_c', 'sports', 'tvsports', 'exercise', 'dining', 'museums',
           'art', 'hiking', 'gaming', 'clubbing', 'reading', 'tv', 'theater',
           'movies', 'concerts', 'music', 'shopping', 'yoga', 'exphappy',
           'expnum','match_es']

decision = ['match','dec','attr', 'sinc', 'intel', 'fun', 'amb', 'shar', 'like', 'prob','met']

evaluation = ['satis_2', 'length', 'numdat_2']

outcome = ['you_call', 'them_cal', 'date_3', 'numdat_3','num_in_3']
```

导入数据之后我们注意到，在这8378条数据中，有非常多次重复记录，即每个实验对象会和不同人进行多次约会，如果直接进行分析可能
导致偏差的出现，所以需要创建一个新的数据表，其中保证每个试验对象只出现一次。

```python
df = df_full[['iid', 'wave'] + personal + evaluation + outcome].drop_duplicates().copy()
print('====================数据处理完毕====================')
print('数据大小:',df.shape[0],'条')
print('表格列头:',df.shape[1],'列')
df.head()
```

### 2.2.1 热力图

* 男性更倾向于选择有吸引力的女性(attr1_1)
* 女性更倾向于选择有抱负的男性(amb1_1)
* 与最终结果match最相关的不是兴趣爱好，而是对方的决定、态度、以及喜欢程度


```python
# 重新对数据进行整理
dating = pd.read_csv('Speed Dating Data.csv', encoding="ISO-8859-1")
dating_1 = dating.iloc[:, 11:28]
dating_2 = dating.iloc[:, 30:35]
dating_3 = dating.iloc[:, 39:43]
dating_4 = dating.iloc[:, 45:67]
dating_5 = dating.iloc[:, 69:74]
dating_6 = dating.iloc[:, 87:91]
dating_7 = dating.iloc[:, 97:102]
dating_8 = dating.iloc[:, 104:107]
date = pd.concat([dating.iloc[:, 0],dating.iloc[:, 2],dating_1,dating_2,dating_3,dating_4,dating_5,dating_6,dating_7,dating_8], axis=1)
date2 = date.dropna()
date3 = date2.drop(['field', 'from', 'career'], axis=1)

# 画出热力图
plt.subplots(figsize=(20,15))
ax = plt.axes()
ax.set_title("Correlation Heatmap")
corr = date3.corr()
sns.heatmap(corr,xticklabels=corr.columns.values,yticklabels=corr.columns.values)
```

### 2.2.3 匹配结果的探索
* 女性比男性更容易收到下一次约会的邀请，且这一现象是在统计上显著的。
* 大概有26%的情况是由于单一方面的“一厢情愿”而导致的匹配失败。
* 但是以上情况并不是因为缺乏吸引力造成的。

```python
# 不同性别下，对方最终的选择分布情况
sns.set(style="ticks", color_codes=True)
g = sns.FacetGrid(date4, col="gender")
g = g.map(plt.hist, "dec_o")
plt.ticklabel_format(useOffset=False, style='plain')
plt.show()

# 对以上想法进行卡方检验
gender_crosstab = pd.crosstab(index=date4.gender, columns=date4.dec_o)
gender_table = sm.stats.Table(gender_crosstab)
gender_rslt = gender_table.test_nominal_association()
print('卡方检验 p-value：',gender_rslt.pvalue)
```

```python
# 有一方愿意而另一方不愿意的情况
no_love_count = len(date4[(date4['dec_o']==0) & (date4['dec']==1)])
perc_broken_heart = no_love_count / len(date4.index)
print('一厢情愿:',perc_broken_heart*100)
```

# 第3部分 建模

1. 自变量的选择：
    * 根据以上分析，吸引力、风趣度、喜欢程度、兴趣相关度这几个因素对最终决策有较大影响
    * 自变量：'dec','dec_o','attr','attr_o','fun','fun_o','like','like_o','int_corr'
    * 因变量：'match'
2. 模型选择：
    * 决策树
    * Boosting模型
    * 随机森林
    * SVM模型

```python
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor, export_graphviz
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn import metrics
import graphviz
```

```python
# 分割训练集、测试集
X = date4[['attr','attr_o','fun','fun_o','like','like_o','int_corr']]
y = date4['match']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

my_feature_name = ['attr','attr_o','fun','fun_o','like','like_o','int_corr']
```

```python
# 决策树

model = DecisionTreeClassifier(criterion = 'entropy',random_state=1,max_depth=3)
clf_model = model.fit(X_train,y_train)
predict_train_clf = clf_model.predict(X_train)
predict_test_clf = clf_model.predict(X_test)
print('Training Accuracy:', metrics.accuracy_score(y_train, predict_train_clf))
print('Validation Accuracy:', metrics.accuracy_score(y_test, predict_test_clf))
print('参数权重：',[*zip(my_feature_name,clf_model.feature_importances_)])

# 画出决策树
dot_data = tree.export_graphviz(clf_model
                                ,feature_names = my_feature_name
                                ,class_names=["不匹配","匹配"]
                                ,filled=True
                                ,rounded=True
                               )
graph = graphviz.Source(dot_data)
graph
```

### 决策树模型总结：
1. 决策树模型在训练数据集上的正确率达到了83.88%，在验证数据集上的正确率达到了83.48%
2. 决策树模型认为最重要的参数有2个，即双方相互的喜欢程度，分别占了0.5137和0.4188，基本对结果起到了决定性作用
3. 从上面决策树的可视化图形中可以看到，最终匹配的条件为：
   * 你喜欢对方的程度大于6.25
   * 对方喜欢你的程度大于6.25
   * 对方认为你的吸引力大于7.75

```python
# Boosting 模型：xgboost模型
model = GradientBoostingClassifier()
xgb_model = model.fit(X_train, y_train)
predict_train_xgb = xgb_model.predict(X_train)
predict_test_xgb = xgb_model.predict(X_test)
print('Training Accuracy:', metrics.accuracy_score(y_train, predict_train_xgb))
print('Validation Accuracy:', metrics.accuracy_score(y_test, predict_test_xgb))
print('参数权重：',[*zip(my_feature_name,xgb_model.feature_importances_)])
```

### XGBoost模型总结：
1. XGBoost模型在训练数据集上的正确率达到了86.69%，在验证数据集上的正确率达到了84.51%
2. XGBoost模型对这些特征权重的分配比较均匀，其中最重要的依然是双方相互的喜欢程度，分别占了0.2672和0.2502，其次重要的是双方相互的吸引力，分别占了0.1408和0.1603

```python
# 随机森林模型
model = RandomForestClassifier(random_state=0)
rf_model = model.fit(X_train, y_train)
predict_train_rf = rf_model.predict(X_train)
predict_test_rf = rf_model.predict(X_test)
print('Training Accuracy:', metrics.accuracy_score(y_train, predict_train_rf))
print('Validation Accuracy:', metrics.accuracy_score(y_test, predict_test_rf))
print('参数权重：',[*zip(my_feature_name,rf_model.feature_importances_)])
```

### 随机森林模型总结：
1. 随机森林模型在训练数据集上的正确率达到了99.98%，在验证数据集上的正确率达到了84.26%，是正确率最高的模型！
2. 随机森林模型对特征权重的分配更加均匀，没有任何一个特征是对结果起到主导作用，也比较符合现实中的情况，双方相互匹配
   不可能只看喜欢程度或吸引力，还是要综合权衡一个人的各个方面，其中单一特征权重最大的是兴趣相关度，权重为0.2681，在现实中也正是
   具有相同爱好的人更容易在一起。

```python
# SVM模型
svm_model = SVC(kernel = "linear").fit(X_train, y_train)
predict_train_svm = svm_model.predict(X_train)
predict_test_svm = svm_model.predict(X_test)
print('Training Accuracy:', metrics.accuracy_score(y_train, predict_train_svm))
print('Validation Accuracy:', metrics.accuracy_score(y_test, predict_test_svm))
print('参数权重：',svm_model.coef_)
```

### SVM模型总结：
1. SVM模型在训练数据集上的正确率达到了82.51%，在验证数据集上的正确率达到了82.55%
2. SVM模型对参数权重的分配主要放在了双方的吸引力程度上，导致权重并不均衡，所以准确率略低

### 模型综合对比
1. 随机森林模型在这四种模型中表现最好，在训练数据集上的正确率达到了99.98%，在验证数据集上的正确率达到了84.26%；
2. 从特征权重角度看，模型只有把这些特征进行综合分析，才能得出较好的预测结果。
