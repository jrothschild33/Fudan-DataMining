### 问题1. 读入市长电话训练集 train_set.csv 和测试集 test_set.csv

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression as LR
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import linear_model
from sklearn.metrics import roc_curve, auc
%matplotlib inline
plt.rcParams['font.sans-serif']=['SimHei']
```

```python
lines = np.loadtxt('train_set.csv', delimiter=',', dtype='str',encoding='utf-8')
print('训练数据集大小:',lines.shape[0]-1)
print('投诉词汇:',lines.shape[1]-1)
print('词汇列表:',lines[0][1:])
```

```python
# 去掉数据中的引号(“)
mycol = np.array([x.strip('"') for x in lines[0]])
myidx = np.array([x.strip('"') for x in lines[:,0]])
lines[0] = mycol
lines[:,0] = myidx
```

```python
data = pd.DataFrame(lines[1:],columns=lines[0])
data.columns.name = "id"
data.head()
```

```python
data2 = pd.DataFrame(lines[1:,1:],columns=lines[0,1:],index=lines[1:,0],dtype='int32')
data2.columns.name = "单位名称"
```

### 问题2.1 统计训练集中各个政府单位接到的市民投诉量，并绘制柱状图（按照投诉量降序）

```python
mydata1 = pd.DataFrame(data2.index)
# 统计索引列中的项目计数
result1 = mydata1[0].value_counts()
# 将结果转换为列表，作为柱状图的x轴
result1_x = result1.index.to_list()
# 统计数量作为柱状图的y轴
result1_y = [i for i in result1]
# 绘制柱状图
plt.figure(figsize=(13, 4))
plt.subplot(111)
x = result1_x
y = result1_y
plt.bar(x,y,width=0.5)
plt.title("柱状图:各个政府单位接到的市民投诉量",fontsize='large')
plt.xlabel('图2-1 各个政府单位接到的市民投诉量',fontsize='medium')
plt.ylabel('市民投诉量',fontsize='medium')
plt.show()
```

### 问题2.2 按政府部门，根据词汇出现的频数绘制分布直方图

```python
# 统计各政府部门词汇出现的总频率，并绘制直方图
gov_depart = ['市水务集团', '市供热公司', '市运输管理局', '市燃气集团', '市公交集团', '市房地集团', '市供电公司']
word_cnt = []
for i in gov_depart:
    temp = data2.loc[i,:]
    word_cnt.append(temp.sum().sum())
word_cnt
```

```python
# 绘制柱状图
plt.figure(figsize=(13, 4))
plt.subplot(111)
x = gov_depart
y = word_cnt
plt.bar(x,y,width=0.5,color='g')
plt.title("直方图:各个政府单位词汇出现的频数分布",fontsize='large')
plt.xlabel('图2-2 各个政府单位词汇出现的频数分布直方图',fontsize='medium')
plt.ylabel('市民投诉量',fontsize='medium')
plt.show()
```

### 问题3 对每一个政府部门，统计每条投诉信息的词汇量，按照政府部门来绘制箱线图，尝试解读这个箱线图呈现的现象；

```python
temp_data = []
for i in gov_depart:
    temp = data2.loc[i,:]
    temp["行总和"] = temp.apply(lambda x:x.sum(),axis =1)
    temp_ = temp["行总和"].values
    temp_data.append(temp_)

temp_data = np.array(temp_data)

# 画出箱线图
plt.figure(figsize=(13,10))
plt.subplot(111)
labels = gov_depart
plt.boxplot([i for i in temp_data],labels=labels)
plt.title("各政府部门投诉信息词汇量箱线图",fontsize='large')
plt.xlabel('图3 投诉信息词汇量箱线图',fontsize='medium')
plt.ylabel('投诉信息词汇量',fontsize='medium')
plt.show()
```

从上述箱线图中可以看出，市房地集团的中位数最大，说明该部门接到投诉词汇量是较大的，但从投诉数量上来看该部门排名倒数第二，这也反映了现实
中房地产业务问题是无法通过较少词汇描述清楚的，和实际情况相符；市水务集团、市公交集团的词汇量中位数时最低的，虽然市水务集团接到的投诉
总量最大，但是看出来每次投诉的词汇量较少。从离群值方面来看，市运输管理局的离群值最多，说明该部门接到的问题差异较大，而市房地集团、
市供电公司的离群值较少，说明这两个部门遇到的问题从词汇量上看差异不大。

### 问题5.1 用朴素贝叶斯方法对市政电话文本进行分类，计算混淆矩阵并可视化

```python
import seaborn as sns
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split

X_train = lines[1:,1:]
Y_train = np.array(data.iloc[:,0].map({'市水务集团':0, '市供热公司':1, '市运输管理局':2, '市燃气集团':3, '市公交集团':4, '市房地集团':5, '市供电公司':6})).reshape(-1,1)
X_train = np.array(X_train,dtype='float64')
Y_train = np.array(Y_train,dtype='float64')

# 划分训练集、测试集
train_X,val_X, train_y, val_y = train_test_split(X_train, Y_train,test_size=0.3)

# 拟合模型
model = GaussianNB()
model.fit(train_X, train_y)
print(model)

# 进行验证
expected = val_y
predicted = model.predict(val_X)

# 输出分类信息
print('分类信息：\n',metrics.classification_report(expected, predicted))

# 输出混淆矩阵信息
label = list(set(Y_train.flatten()))
sns.set()
f,ax=plt.subplots()
C2= confusion_matrix(expected, predicted, labels=label)
sns.heatmap(C2,annot=True,ax=ax)
ax.set_title('confusion matrix')
ax.set_xlabel('predict')
ax.set_ylabel('true')
```

### 问题5.2 使用该模型对测试集电话文本进行预测，计算混淆矩阵并可视化

```python
# 导入测试集数据
lines = np.loadtxt('test_set.csv', delimiter=',', dtype='str',encoding='utf-8')
print('测试数据集大小:',lines.shape[0]-1)
print('投诉词汇:',lines.shape[1]-1)
print('词汇列表:',lines[0][1:])

# 去掉数据中的引号(“)
mycol = np.array([x.strip('"') for x in lines[0]])
myidx = np.array([x.strip('"') for x in lines[:,0]])
lines[0] = mycol
lines[:,0] = myidx

# 将数据转换为DataFrame
data = pd.DataFrame(lines[1:],columns=lines[0])
data.columns.name = "id"

# 提取features和labels
X_test = lines[1:,1:]
Y_test = np.array(data.iloc[:,0].map({'市水务集团':0, '市供热公司':1, '市运输管理局':2, '市燃气集团':3, '市公交集团':4, '市房地集团':5, '市供电公司':6})).reshape(-1,1)
X_test = np.array(X_test,dtype='float64')
Y_test = np.array(Y_test,dtype='float64')

# 进行验证
expected = Y_test
predicted = model.predict(X_test)

# 输出分类信息
print('分类信息：\n',metrics.classification_report(expected, predicted))

# 输出混淆矩阵信息
label = list(set(Y_test.flatten()))
sns.set()
f,ax=plt.subplots()
C2= confusion_matrix(expected, predicted, labels=label)
sns.heatmap(C2,annot=True,ax=ax)
ax.set_title('confusion matrix')
ax.set_xlabel('predict')
ax.set_ylabel('true')
```

本案例中先将各个政府部门分为7类，用数字0~6标记代替，然后利用train_set中的数据划分70%训练朴素贝叶斯分类器模型，并用剩余30%部分进行验证，可见
在570个验证数据集上的准确率为87%，而在测试集test_set中的表现提升至了91%，可见朴素贝叶斯分类器的效果较佳，其中准确率最高的是对'市运输管理局'、
'市公交集团'、'市供电公司'的预测，准确率达到了100%，但也可能是由于数据集过小，不太具有代表性，作为被投诉量第一的'市水务集团'，分类正确率达到了
89%，而效果最差的是'市房地集团'，准确率仅67%。
